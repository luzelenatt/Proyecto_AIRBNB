#### 2.1.3 Inconsistencias en los tipos de datos

```python
print("\nTipos de datos actuales:")
print(df_original.dtypes)
```

```python
# Comparar tipos de datos actuales con el diccionario de datos proporcionado
# Diccionario de datos
data_dict = {
    "id": "int",
    "name": "str",
    "host_id": "int",
    "host_name": "str",
    "neighbourhood_group": "str",
    "neighbourhood": "str",
    "latitude": "float",
    "longitude": "float",
    "room_type": "str",
    "price": "float",
    "minimum_nights": "int",
    "number_of_reviews": "int",
    "last_review": "datetime",
    "reviews_per_month": "float",
    "calculated_host_listings_count": "int",
    "availability_365": "int",
    "number_of_reviews_ltm": "int",
    "license": "str"
}
```

```python
# Chequeo de consistencia
inconsistencias = []
for column, expected_type in data_dict.items():
    if column in df_original.columns:
        actual_type = str(df_original[column].dtype)
        if expected_type == "str" and actual_type == "object":
            actual_type = "str"
        if expected_type == "datetime" and actual_type.startswith("datetime"):
            actual_type = "datetime"
        if actual_type != expected_type:
            inconsistencias.append({
                "Columna": column,
                "Esperado": expected_type,
                "Actual": actual_type,
                "Consistencia": "CON OBSERVACIÓN"
            })
        else:
            inconsistencias.append({
                "Columna": column,
                "Esperado": expected_type,
                "Actual": actual_type,
                "Consistencia": "SIN OBSERVACIÓN"
            })
    else:
        inconsistencias.append({
            "Columna": column,
            "Esperado": expected_type,
            "Actual": "No encontrada",
            "Consistencia": "CON OBSERVACIÓN"
        })
```

```python
# Convertir las inconsistencias a un DataFrame para mostrar en forma de tabla
inconsistencias_df = pd.DataFrame(inconsistencias)
```

```python
print("\nInconsistencias encontradas en los tipos de datos:")
print(inconsistencias_df)
```

### 2.2 Limpiar los datos

#### 2.2.1 Eliminar o imputar valores nulos

```python
# Crear una copia del DataFrame original para realizar la limpieza
df_cleaned = df_original.copy()
```

```python
# Imputación de valores nulos en 'reviews_per_month' con la media
df_cleaned['reviews_per_month'].fillna(df_cleaned['reviews_per_month'].mean(), inplace=True)
```

```python
df_cleaned
```

```python
# Convertir 'last_review' a datetime antes de imputar
df_cleaned['last_review'] = pd.to_datetime(df_cleaned['last_review'], errors='coerce')

# Calcular la mediana de 'last_review' y usarla para imputar los valores nulos
median_last_review = df_cleaned['last_review'].median()
df_cleaned['last_review'].fillna(median_last_review, inplace=True)
```

```python
df_cleaned
```

```python
# Imputación de valores nulos en 'neighbourhood_group' con 'Unknown'
df_cleaned['neighbourhood_group'].fillna('Unknown', inplace=True)
df_cleaned
```

```python
# Imputación de valores nulos en 'price' con la media
df_cleaned['price'].fillna(df_cleaned['price'].mean(), inplace=True)
df_cleaned
```

```python
# Imputación de valores nulos en 'license' con 'No license'
df_cleaned['license'].fillna('No license', inplace=True)
df_cleaned
```

#### 2.2.2 Eliminar filas duplicadas

```python
initial_row_count = df_cleaned.shape[0]
df_cleaned.drop_duplicates(inplace=True)
final_row_count = df_cleaned.shape[0]
duplicates_removed = initial_row_count - final_row_count
```

```python
df_cleaned
```

#### 2.2.3 Convertir tipos de datos incorrectos

```python
df_cleaned['id'] = df_cleaned['id'].astype(int)
df_cleaned['host_id'] = df_cleaned['host_id'].astype(int)
df_cleaned['price'] = df_cleaned['price'].astype(float)
df_cleaned['minimum_nights'] = df_cleaned['minimum_nights'].astype(int)
df_cleaned['number_of_reviews'] = df_cleaned['number_of_reviews'].astype(int)

# Convertir 'last_review' a datetime especificando el formato de fecha o usando dayfirst=True
df_cleaned['last_review'] = pd.to_datetime(df_cleaned['last_review'], format='%d/%m/%Y', errors='coerce')

df_cleaned['reviews_per_month'] = df_cleaned['reviews_per_month'].astype(float)
df_cleaned['calculated_host_listings_count'] = df_cleaned['calculated_host_listings_count'].astype(int)
df_cleaned['availability_365'] = df_cleaned['availability_365'].astype(int)
df_cleaned['number_of_reviews_ltm'] = df_cleaned['number_of_reviews_ltm'].astype(int)
```

```python
# Mostrar información del DataFrame después de la limpieza
print("\nInformación del DataFrame después de la limpieza:")
print(df_cleaned.info())
```

```python
# Verificar que los valores nulos han sido eliminados o imputados
print("\nValores nulos por columna después de la limpieza:")
print(df_cleaned.isnull().sum())
```

```python
# Guardar el DataFrame limpio en un nuevo archivo CSV
df_cleaned.to_csv('listings_cleaned.csv', index=False)
```

```python
# Crear reporte de datos eliminados o modificados
modifications_report = {
    "Filas originales": df_original.shape[0],
    "Filas finales": df_cleaned.shape[0],
    "Filas eliminadas": df_original.shape[0] - df_cleaned.shape[0],
    "Duplicados eliminados": duplicates_removed,
    "Valores nulos imputados": (df_original.isnull().sum() - df_cleaned.isnull().sum()).sum()
}
```

## 3. Transformación de Datos

### 3.1 Normalizar y estandarizar columnas

#### 3.1.1 Convertir nombres de columnas a minúsculas

```python
df_cleaned.columns = [col.lower() for col in df_cleaned.columns]
df_cleaned
```

#### 3.1.2 Renombrar columnas para mejorar la claridad

```python
df_cleaned.rename(columns={
    'id': 'id',
    'name': 'nombre',
    'host_id': 'id_anfitrion',
    'host_name': 'nombre_anfitrion',
    'neighbourhood_group': 'grupo_vecindario',
    'neighbourhood': 'vecindario',
    'latitude': 'latitud',
    'longitude': 'longitud',
    'room_type': 'tipo_habitacion',
    'price': 'precio',
    'minimum_nights': 'noches_minimas',
    'number_of_reviews': 'numero_resenas',
    'last_review': 'ultima_resena',
    'reviews_per_month': 'resenas_por_mes',
    'calculated_host_listings_count': 'conteo_listados_anfitrion',
    'availability_365': 'disponibilidad_365',
    'number_of_reviews_ltm': 'numero_resenas_ltm',
    'license': 'licencia'
}, inplace=True)
```

```python
# Crear un nuevo DataFrame de transformación
df_transformation = df_cleaned.copy()
df_transformation
```

### 3.2 Crear nuevas columnas

```python
# Crear categorías de precios (bajo, medio, alto)
price_bins = [0, 50, 150, float('inf')]
price_labels = ['Bajo', 'Medio', 'Alto']
df_transformation['categoria_precio'] = pd.cut(df_transformation['precio'], bins=price_bins, labels=price_labels)
```

```python
df_transformation
```

```python
# Crear categorías de disponibilidad (baja, media, alta)
availability_bins = [0, 100, 200, 365]
availability_labels = ['Baja', 'Media', 'Alta']
df_transformation['categoria_disponibilidad'] = pd.cut(df_transformation['disponibilidad_365'], bins=availability_bins, labels=availability_labels)

# Asegurarse de que 'sin_categoria' esté en los valores posibles
df_transformation['categoria_disponibilidad'] = df_transformation['categoria_disponibilidad'].cat.add_categories('Sin Categoria')
df_transformation['categoria_disponibilidad'].fillna('Sin Categoria', inplace=True)
df_transformation
```

### 3.3 Realizar transformaciones de datos adicionales según sea necesario

```python
# Extraer mes y año de la última reseña con nombres de meses
df_transformation['mes_ultima_resena'] = df_transformation['ultima_resena'].dt.strftime('%B')
df_transformation['ano_ultima_resena'] = df_transformation['ultima_resena'].dt.year
df_transformation
```

```python
# Simplificar tipo de habitación
room_type_mapping = {
    'Entire home/apt': 'Entero',
    'Private room': 'Privado',
    'Shared room': 'Compartido',
    'Hotel room': 'Hotel'
}
df_transformation['tipo_habitacion_simplificado'] = df_transformation['tipo_habitacion'].map(room_type_mapping)
df_transformation
```

```python
# Crear rango basado en el promedio de reseñas por mes
reviews_per_month_bins = [0, 0.5, 1.5, float('inf')]
reviews_per_month_labels = ['Bajo', 'Medio', 'Alto']
df_transformation['rango_resenas_por_mes'] = pd.cut(df_transformation['resenas_por_mes'], bins=reviews_per_month_bins, labels=reviews_per_month_labels)
df_transformation
```

```python
# Categorizar el número de listados del anfitrión
host_listings_bins = [0, 5, 20, float('inf')]
host_listings_labels = ['Pocos', 'Moderado', 'Muchos']
df_transformation['categoria_listados_anfitrion'] = pd.cut(df_transformation['conteo_listados_anfitrion'], bins=host_listings_bins, labels=host_listings_labels)
df_transformation
```

```python
# Mostrar información del DataFrame de transformación
print("\nInformación del DataFrame de transformación:")
print(df_transformation.info())
```

```python
# Guardar el DataFrame de transformación en un nuevo archivo CSV
df_transformation.to_csv('listings_transformed.csv', index=False)
```

```python
# Detección de valores atípicos en columnas numéricas usando el método IQR
print("\nValores atípicos en columnas numéricas:")
for column in numerical_columns:
    Q1 = df_transformation[column].quantile(0.25)
    Q3 = df_transformation[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_transformation[(df_transformation[column] < (Q1 - 1.5 * IQR)) | (df_transformation[column] > (Q3 + 1.5 * IQR))]
    print(f"{column}: {len(outliers)} valores atípicos")
```

```python
# Revisión de correlaciones entre las variables numéricas
print("\nCorrelaciones entre variables numéricas:")
correlation_matrix = df_transformation[numerical_columns].corr()
correlation_matrix
```
